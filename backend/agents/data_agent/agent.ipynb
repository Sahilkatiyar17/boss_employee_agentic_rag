```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv(r'F:\sahil\2025-2026\Project_DS\boss_employee_agentic_rag\backend\data\Sales-Export_2019-2020.csv')

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')

# Calculate profit
df['profit'] = df['order_value_EUR'] - df['cost']

# Basic statistics
print(df.describe())

# Visualization 1: Total order value by country
plt.figure(figsize=(10,6))
sns.barplot(x='country', y='order_value_EUR', data=df, estimator=np.sum)
plt.title('Total Order Value by Country')
plt.xticks(rotation=45)
plt.show()

# Visualization 2: Monthly sales trend
df_monthly = df.resample('M', on='date')['order_value_EUR'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(x='date', y='order_value_EUR', data=df_monthly)
plt.title('Monthly Total Order Value')
plt.show()

# Visualization 3: Distribution of order values by category
plt.figure(figsize=(10,6))
sns.boxplot(x='category', y='order_value_EUR', data=df)
plt.title('Order Value Distribution by Category')
plt.xticks(rotation=45)
plt.show()

# Visualization 4: Top sales reps by total sales
top_reps = df.groupby('sales_rep')['order_value_EUR'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_reps.index, y=top_reps.values)
plt.title('Top 10 Sales Reps by Total Sales')
plt.xticks(rotation=45)
plt.show()

# Visualization 5: Cost vs Order Value
plt.figure(figsize=(10,6))
sns.scatterplot(x='cost', y='order_value_EUR', hue='category', data=df)
plt.title('Cost vs Order Value by Category')
plt.show()
```
